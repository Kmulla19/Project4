```python
import pandas as pd
import numpy as np

from textblob import TextBlob
import nltk
```

```python
df = pd.read_csv('data/tweets_2017.csv', encoding='latin1')
df.rename(columns={"Tweet Text": "body"}, inplace=True)
df.tail()
```

```python
df = df.astype('str')

apple_words = ['apple', 'ipad', 'iphone', 'mac', 'ios']
google_words = ['google', 'android', 'pixel']

def is_apple(x):
    for i in apple_words:
            if i.lower() in x.lower():
                return 'apple'
            else:
                continue

    
def is_google(x):
    for i in google_words:
            if i.lower() in x.lower():
                return 'google'
            else:
                continue
```

```python
df['company'] = df.body.apply(is_apple)
df['google'] = df.body.apply(is_google)
df['company'] = df['company'].combine_first(df['google'])
df.drop(columns=['google'], inplace=True)


print(df.company.value_counts())
df.sample(50)
```

```python
tweets = pd.read_csv('data/tweet_tweet.csv')

tweets.rename(
    columns={
        'tweet_text': 'body',
        'emotion_in_tweet_is_directed_at': 'product',
        'is_there_an_emotion_directed_at_a_brand_or_product': 'target'
    },
    inplace=True
)


sentimental_dict = {
    'No emotion toward brand or product': 'neutral',
    'Negative emotion': 'negative',
    'Positive emotion': 'positive'
}
```

```python
tweets = tweets.astype('str')

# function to get tweet polarity
def polaritizer(tweet):
    # polarize each tweet
    return TextBlob(tweet).polarity

# add new polarity column
tweets['polarity'] = tweets.body.apply(polaritizer)

# create new column with positive, negative, neutral
def sentiment(polarity):
    if polarity < -0.25:
        return 'negative'
    elif polarity > 0.15:
        return 'positive'
    else:
        return 'neutral'
    
tweets['sentiment'] = tweets.polarity.apply(sentiment)
tweets.target = tweets.target.map(sentimental_dict)
print(tweets.target.value_counts())
print(tweets.sentiment.value_counts())
```

```python
# difference_filter = tweets[tweets.target != tweets.sentiment]
tweet_diff = tweets.loc[tweets.target != tweets.sentiment]
```

```python
df = pd.read_csv('data/tweet_tweet.csv')

import re
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')

df = df.reset_index()
df.rename(columns={'is_there_an_emotion_directed_at_a_brand_or_product': 'body'})
#create a new data frame with "id" and "comment" fields
df_subset = df[['index', 'body']].copy()

#data clean-up
#remove all non-aphabet characters
df_subset['body'] = df_subset['body'].str.replace("[^a-zA-Z#]", " ")

#covert to lower-case
df_subset['body'] = df_subset['body'].str.casefold()
print (df_subset.head(10))

# set up empty dataframe for staging output
df1 = pd.DataFrame()
df1['index'] = ['99999999999']
df1['sentiment_type']='NA999NA'
df1['sentiment_score']=0

print('Processing sentiment analysis...')
sid = SentimentIntensityAnalyzer()
t_df = df1
for index, row in df_subset.iterrows():
    scores = sid.polarity_scores(row[1])

    for key, value in scores.items():
        temp = [key, value, row[0]]
        df1['index'] = row[0]
        df1['sentiment_type'] = key
        df1['sentiment_score'] = value
        t_df = t_df.append(df1)

#remove dummy row with row_id = 99999999999
t_df_cleaned = t_df[t_df.index != '99999999999']

#remove duplicates if any exist
t_df_cleaned = t_df_cleaned.drop_duplicates()

# only keep rows where sentiment_type = compound
t_df_cleaned = t_df[t_df.sentiment_type == 'compound']

print(t_df_cleaned.head(10))

#merge dataframes
df = pd.merge(df, t_df_cleaned, on='index', how='inner')

df.head(24)
```

```python
# create new column with positive, negative, neutral
def sentiment(polarity):
    if polarity < -0.5:
        return 'negative'
    elif polarity > 0.5:
        return 'positive'
    else:
        return 'neutral'
    
df['sentiment'] = df.sentiment_score.apply(sentiment)
df['body'] = df['body'].map(sentimental_dict)
df_diff = df.loc[df['body'] != df.sentiment]
df_diff
```
